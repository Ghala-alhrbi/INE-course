```python
#assignment1 list to dicts
def convert_list_to_dicts(players_list):
    users = []
    for user in players_list:
        user1 = {

            'number': user[0],
            'position': user[1],
            'name': user[2],
            'date_of_birth': user[3],
            'caps': user[4],
            'club': user[5],
            'country': user[6],
            'club_country': user[7],
            'year': user[8]
        }
        users.append(user1)
    return users
convert_list_to_dicts(SQUADS_DATA)
        
    
```

```python
#assignment2 group data by position

def players_by_position(squads_list):
    positions = {}
    for player in squads_list:
        if player[1] not in positions:
            positions[player[1]] = []
    for player in squads_list:
        player_dict = {
            'number': player[0],
            'position': player[1],
            'name': player[2],
            'date_of_birth': player[3],
            'caps': player[4],
            'club': player[5],
            'country': player[6],
            'club_country': player[7],
            'year': player[8]
        }
        position = player[1]
        positions[position].append(player_dict)
    return positions
        

players = players_by_position(SQUADS_DATA)

for a,b in players.items():
    print(a,b,'\n')
```

```python
#assignment3 group data by country and position

def players_by_country_and_position(squads_list):
    countries = {}

    for player in squads_list:
        country = player[6]  
        position = player[1]  
        if country not in countries:
            countries[country] = {}
        if position not in countries[country]:
            countries[country][position] = []
            
        player_dict = {
            'number': player[0],
            'position': player[1],
            'name': player[2],
            'date_of_birth': player[3],
            'caps': player[4],
            'club': player[5],
            'country': player[6],
            'club_country': player[7],
            'year': player[8]
        }
        
        countries[country][position].append(player_dict)
    return countries
        

players = players_by_country_and_position(SQUADS_DATA)

for a,b in players.items():
    print(a,b ,'\n')
```
